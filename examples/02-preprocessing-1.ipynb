```python
# Function to process output/label column
def process_lab_results(series):
    # Step 1: Identify intermediate values that do not repeat and change them to None
    # Values between two laboratory results
    shifted_forward = series.shift(-1)
    shifted_backward = series.shift(1)
    mask_intermediate = (series != shifted_forward) & (series != shifted_backward)
    series = series.where(~mask_intermediate, None)
    
    # Step 2: Retain only the first occurrence of consecutive repeated values
    # The true laboratory value on correct timestamp
    mask_first_occurrence = series != series.shift()
    series = series.where(mask_first_occurrence, None)
    
    return series
```

```python
df['FLUIDEZ_LAB_processed'] = process_lab_results(df['FLUIDEZ_LAB'])

# After previous function, the first row can be a intermediate value
# Step 1: Delete the first row
df = df.iloc[1:]
```

```python
# Function to create moving averages for not RNN models
def create_moving_average_data_with_valid_labels(df, window):
    x, y = [], []

    # Iterate through the DataFrame to find valid labels first
    for i in range(window, len(df)):
        label = df.iloc[i, -1]
        if not np.isnan(label):  # Only proceed if the label is not NaN
            features = df.iloc[i - window + 1:i + 1, :-1].mean(axis=0).values  # Get the moving average features
            x.append(features)
            y.append(label)
            
    return np.array(x), np.array(y)

WINDOW = 5
x, y = create_moving_average_data_with_valid_labels(df2, WINDOW)
```

```python
# Remove data rows with not a number values
# Find indices where NaN values are present in x
nan_indices = [i for i in range(x.shape[0]) if np.isnan(x[i]).any()]

# Remove corresponding rows in x and y
x_cleaned = np.delete(x, nan_indices, axis=0)
y_cleaned = np.delete(y, nan_indices, axis=0)

print(f"Original x shape: {x.shape}")
print(f"Original y shape: {y.shape}")
print(f"Cleaned x shape: {x_cleaned.shape}")
print(f"Cleaned y shape: {y_cleaned.shape}")
```

```python
# Remove data rows with zeros
# Specify which columns to remove zeros
columns_to_check = ['T_QAV', 'T_DL', 'T_DP', 'REND_QAV', 'REND_DL', 'REND_DP', 'T_RC_DL',
       'T_RC_DP', 'T_ZONA_FLASH']

# Get the list of column indexes
column_indexes = [df2.columns.get_loc(col) for col in columns_to_check]

# Check the dimensions of x_cleaned and apply the appropriate logic
if x_cleaned.ndim == 3:
    # Find indices where zero values are present in the specified columns for 3D data
    zero_indices = []
    for i in range(x_cleaned.shape[0]):
        if np.any(x_cleaned[i, :, column_indexes] == 0):
            zero_indices.append(i)
elif x_cleaned.ndim == 2:
    # Find indices where zero values are present in the specified columns for 2D data
    zero_indices = np.any(x_cleaned[:, column_indexes] == 0, axis=1).nonzero()[0]
else:
    raise ValueError("x_cleaned should be either 2D or 3D")

# Remove corresponding rows in x and y
x_cleaned2 = np.delete(x_cleaned, zero_indices, axis=0)
y_cleaned2 = np.delete(y_cleaned, zero_indices, axis=0)

print(f"Original x shape: {x_cleaned.shape}")
print(f"Original y shape: {y_cleaned.shape}")
print(f"Cleaned x shape: {x_cleaned2.shape}")
print(f"Cleaned y shape: {y_cleaned2.shape}")
```

```python
# Remove data rows with infinite values
# Find indices where inf values are present in x
inf_indices = [i for i in range(x_cleaned2.shape[0]) if np.isinf(x_cleaned2[i]).any()]

# Remove corresponding rows in x and y
x_cleaned3 = np.delete(x_cleaned2, inf_indices, axis=0)
y_cleaned3 = np.delete(y_cleaned2, inf_indices, axis=0)

print(f"Original x shape: {x_cleaned2.shape}")
print(f"Original y shape: {y_cleaned2.shape}")
print(f"Cleaned x shape: {x_cleaned3.shape}")
print(f"Cleaned y shape: {y_cleaned3.shape}")
```

```python
# Calculating outliers from 'y' array and removing respectively rows
# SCALE = 1.5  # = 2.7 standard deviations
SCALE = 1.7  # = 3 standard deviations
# SCALE = 3.21 # = 5 standard deviations
# SCALE = 6.91 # = 10 standard deviations

Q1 = np.percentile(y, 25)
Q3 = np.percentile(y, 75)
IQR = Q3 - Q1  # interquartile range

lower_bound = Q1 - SCALE * IQR
upper_bound = Q3 + SCALE * IQR

# Determine the outliers in y
outliers_mask = (y_cleaned3 < lower_bound) | (y_cleaned3 > upper_bound)

# Filter out the corresponding rows in both x and y
x_filtered = x_cleaned3[~outliers_mask]
y_filtered = y_cleaned3[~outliers_mask]

# Output the shapes of the filtered arrays for verification
print(f'Shape of x_filtered: {x_filtered.shape}')
print(f'Shape of y_filtered: {y_filtered.shape}')
```
